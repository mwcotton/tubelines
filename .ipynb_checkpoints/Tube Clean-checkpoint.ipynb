```python
max_time = 1000
tests = 200

plt.figure(figsize=(8,8))

index = 0

for linename in ['Central Line', 'Waterloo & City Line', 'Northern Line']:

    line = line_objects[linename]
    
    moving_from_line = []

    for _ in range(tests):
        start = np.random.choice(line.line_list)
        moving_from_line.extend(time_track_ensemble(start-1, 1, adj_mat, max_time))

    sim_dists = {}

    for i in range(0, max_time):
        sim_dists[i] = [line.dist_to_node(track[i]) for track in moving_from_line]
    
    means = []
    stds =[]

    for i in sim_dists:
        means.append(np.mean(sim_dists[i]))
        stds.append(np.std(sim_dists[i]))

    plt.plot(range(0, max_time), means, c='#{}'.format(cols[linename]), label=linename)
    
    def func(x, a):
         return a * np.sqrt(x)

    popt, pcov = sp.optimize.curve_fit(func, range(0, max_time)[0:1000], means[0:1000])


    plt.plot(np.array(range(0, max_time)), func(np.array(range(0, max_time)), *popt), '--', c='#{}'.format(cols[linename]), label=('y ={:.2f}x^0.5  ({:.2f}e-6)'.format(*popt, 1000000*pcov[0][0])))
    
plt.ylabel('Mean Distance from Line')
plt.xlabel('Time')
plt.title('Growing Distance from Line')
plt.legend()
plt.savefig('1000 spec start mean changes {}{}.png'.format(tests, max_time), dpi=400)
```

```python
def return_to_line(start, line, connect_mat):
    #start_loc = matrix form
    
    next_loc = start
    track = []

    track.append(next_loc)

    #so that it is never equal to the start location or is on the line
    paths = (np.nonzero(connect_mat[start])[0])


    next_loc = np.random.choice(paths)

    #print(np.sort(line.line_list))
    
    while not line.line_list.count(next_loc+1):

        #print(line.line_list.count(next_loc))
        
        location = int(next_loc)
        paths = (np.nonzero(connect_mat[location])[0])

        next_loc = np.random.choice(paths)
        track.append(location)

    track.append(next_loc)

    return track
```

```python
trials = 100000

line_return_paths = []

hist_data = {}

fig, axes = plt.subplots(3, 4, figsize=(20,15))

ax_list = axes.flatten()

for linename, ax in zip(line_objects.keys(), ax_list):

    line = line_objects[linename]

    for _ in range(trials):
        line_return_paths.append(return_to_line(np.random.choice(line.line_list)-1, line, adj_mat))

    y = ax.hist([len(path) for path in line_return_paths], alpha=0.2, color='#{}'.format(cols[linename]), label=linename)
    ax.set_yscale('log')
    #ax.set_xlabel('Return Steps')
    hist_data[linename] = y
    
    ax.set_title(linename)
    
plt.savefig('line return times histograms{}'.format(trials), png='400')
```
